# Imports

```python
import os
from dotenv import load_dotenv
from sqlalchemy import create_engine
import pickle
import numpy as np
import geopandas as gpd
from nested_cv.gwr import (
    evaluate_gwr
)
```

```python
load_dotenv()
username = os.getenv("DB_USERNAME")
password = os.getenv("DB_PASSWORD")
```

```python
engine = create_engine(f"postgresql://{username}:{password}@localhost:5432/greenspace")
```

# Set random seed

```python
np.random.seed(42)
```

# Prepare data

## Load data

```python
raw_df = gpd.read_postgis("SELECT * FROM split_full_dataset", engine, "geometry")
```

```python
df = raw_df.copy()
df = df.drop(columns = ["ogc_fid"])
```

# Evaluate model

```python
results = evaluate_gwr(df)
```

# Save output

```python
with open("outputs/model_results/gwr_fe_reduced.pkl", "wb") as f:
    pickle.dump(results, f)
```
