# Imports

```python
import pickle
import numpy as np
import geopandas as gpd
from utils.db_utils import engine
from utils.nested_cv.gwr import (
    evaluate_gwr
)
```

# Set random seed

```python
np.random.seed(42)
```

# Prepare data

```python
raw_df = gpd.read_postgis("SELECT * FROM split_engineered_dataset", engine, "geometry")
```

```python
df = raw_df.copy()
df = df.drop(columns = ["ogc_fid"])
df = df.drop(columns = [col for col in df.columns if "prevalent" in col])
```

# Evaluate model

```python
results = evaluate_gwr(df)
```

# Save output

```python
with open("outputs/model_results/gwr_fe_reduced.pkl", "wb") as f:
    pickle.dump(results, f)
```
