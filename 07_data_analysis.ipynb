# Imports

```python
import numpy as np
import pandas as pd
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

# Load datasets

```python
raw_sar_results = pd.read_csv("datasets/results/test_df_sar.csv")
```

```python
sar_results = raw_sar_results.copy()
```

```python
sar_results = sar_results.dropna(subset=['sar_predictions'])
```

```python
sar_features = sar_results.copy()
sar_labels = sar_results.pop('very_good_health')
sar_predictions = sar_results.pop("sar_predictions")
```

# Calculate performance metrics

```python
rmse = mean_squared_error(sar_labels, sar_predictions)
mae = mean_absolute_error(sar_labels, sar_predictions)
r2 = r2_score(sar_labels, sar_predictions)
```

```python
print(f"RMSE: {rmse :3}")
print(f"MAE: {mae :3}")
print(f"r2: {r2 :3}")
```
