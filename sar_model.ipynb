# Imports

```python
import numpy as np
import pandas as pd
import geopandas as gpd
from pysal.lib import weights
from pysal.explore import esda
import spreg
```

# Load dataset

```python
raw_df = gpd.read_file("datasets/combined/lsoa_greenspace.gpkg")
```

```python
df = raw_df.copy()
```

```python
df = df.drop(columns = ["lsoa", "geometry", "good_health", "fair_health", "bad_health", "very_bad_health", "total_area", "greenspace_area"])
```

# Split data

```python
train_df = df.sample(frac=0.8, random_state=0)
test_df = df.drop(train_df.index)

train_features = train_df.copy()
test_features = test_df.copy()

train_labels = train_features.pop('very_good_health')
test_labels = test_features.pop('very_good_health')
```

# Create OLS regression model

```python
ols_model = spreg.OLS(
    train_labels,
    train_features[['greenspace_proportion', 'imd']],
    name_y = 'very_good_health',
    name_x = ['greenspace_proportion', 'imd'],
)
```

```python
print(ols_model.summary)
```

# Calculate spatial weights

```python
knn = weights.KNN.from_dataframe
```

# Calculate spatial lag

```python
lag_residual = weights.spatial_lag.lag_spatial(knn, ols_model.u)
```
