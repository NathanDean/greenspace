# Imports

```python
import pandas as pd
import geopandas as gpd
```

# Load dataset

```python
raw_df = gpd.read_file("datasets/combined/lsoa_greenspace.gpkg")
```

```python
df = raw_df.copy()
```

```python
df = df.drop(columns = ["lsoa", "geometry", "good_health", "fair_health", "bad_health", "very_bad_health", "total_area", "greenspace_area"])
```

# Create data splits

```python
train_df = df.sample(frac=0.8, random_state=0)
test_df = df.drop(train_df.index)
```

# Save data splits

```python
df.to_csv("datasets/split/complete_df.csv")
train_df.to_csv("datasets/split/train_df.csv")
test_df.to_csv("datasets/split/test_df.csv")
```
