# Imports

```python
import pickle
import numpy as np
import geopandas as gpd
import tensorflow as tf
from utils.db_utils import engine
from utils.nested_cv.ann import (
    evaluate_ann
)
```

# Set random seeds

```python
tf.random.set_seed(42)
np.random.seed(42)
```

# Prepare data

```python
raw_df = gpd.read_postgis("SELECT * FROM split_full_dataset", engine, "geometry")
```

```python
df = raw_df.copy()
df = df.drop(columns = ["ogc_fid"])
```

#  Evaluate model

```python
results = evaluate_ann(df, regularised=True)
```

# Save output

```python
with open("outputs/model_results/regularised_ann_full.pkl", "wb") as f:
    pickle.dump(results, f)
```
